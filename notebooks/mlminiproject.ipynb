```python
import pandas as pd

df = pd.read_csv("30k_cleaning_demo_dataset.csv")
```

```python
# 1. Convert date
df['date'] = pd.to_datetime(df['date'])
```

```python
# 2. Handle missing values

# employee_name → fill with "Unknown"
df['employee_name'] = df['employee_name'].fillna("Unknown")

# customer_id → keep as "Guest" (business logic)
df['customer_id'] = df['customer_id'].fillna("Guest")
```

```python
# 3. Remove redundant columns (optional but recommended)
df = df.drop(['year', 'month', 'day'], axis=1)
```

```python
# 4. Validate calculations (VERY IMPORTANT)

# revenue = quantity * unit_price
df['revenue_check'] = df['quantity'] * df['unit_price']

# profit = revenue - cost
df['profit_check'] = df['revenue'] - df['cost']
```

```python
# 5. Remove inconsistent rows (if mismatch exists)
df = df[df['revenue'] == df['revenue_check']]
df = df[df['profit'] == df['profit_check']]

# drop check columns
df = df.drop(['revenue_check', 'profit_check'], axis=1)
```

```python
df = df.drop_duplicates()
```

```python
# 6. Remove duplicates
df = df.drop_duplicates()
```

```python
# 7. Reset index
df = df.reset_index(drop=True)
```

```python
# Base dataset (cleaned)
df_main = df.copy()
```

```python
#Demand Forecast Dataset
df_demand = df_main.copy()

df_demand = df_demand.drop(columns=[
    'employee_name',
    'employee_id',
    'customer_id'
])
```

```python
#Sales Forecast Dataset
df_sales = df_main.copy()

df_sales = df_sales.drop(columns=[
    'employee_name',
    'employee_id',
    'customer_id'
])
```

```python
#Profit Optimization Dataset

df_profit = df_main.copy()

df_profit = df_profit.drop(columns=[
    'employee_name',
    'employee_id',
    'customer_id'
])
```

```python
#Inventory Optimization Dataset
df_inventory = df_main.copy()

df_inventory = df_inventory.drop(columns=[
    'employee_name',
    'employee_id',
    'customer_id'
])
```
